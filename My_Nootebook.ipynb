### The following code computes the factorial of a number in two ways- recursively and iteratively

````python
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

def factorial_reqursively(num):
    if num == 0 or num == 1:
        return 1
    else:
        return factorial_reqursively(num - 1) * num
    
def factorial_iteratively(num):
    result = 1
    while(num > 1):
        result *= num
        num = num - 1
    return result

for num in range(0,11):

    print("\n" + bcolors.OKBLUE + "Recursive Factorial Result of " + str(num) + "! is: " + bcolors.ENDC, factorial_reqursively(num), "\n\n" +
         bcolors.OKGREEN + "Iterative Factorial Result of " + str(num) + "! is: " + bcolors.ENDC, factorial_iteratively(num))
````
